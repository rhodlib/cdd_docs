# Visualizacion de datos

En este segmento se va a recopilar maneras de visualizar datos utilizando librearias como Matplotlib, Seaborn y Pandas.

La visualizacion de datos nos ayuda a identificar importantes patrones en los datos que pueden ser utilizados para tomar
importantes deciciones en una organizacion.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import math
```

## Diagramas de lineas

```python
"""
Un diagrama de lineas es el primer diagrama que vamos a utilizar, es el mas sencillo de todos. y se utiliza
para mostrar la relacion entre dos grupos de numeros
"""

# Generamos una lista de 20 numeros equisdistante.
x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.plot(x_vals, y_vals)
```

```python
"""
Hay otras maneras de crear este diagrama, utilizando la funcion figure()
esto nos sirve para realizar multiples plots.
"""

fig = plt.figure()
ax = plt.axes()
ax.plot(x_vals, y_vals)
```

```python
"""
Tambien se puede aumentar el tamaño por defecto de un diagrama en matplotlib
para hacerlo se pueden utilizar los rcParams 
"""

plt.rcParams["figure.figsize"] = [8,6] # 8 pulgadas de ancho y 6 pulgadas de alto

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.plot(x_vals, y_vals)
```

## Titulos, Etiquetas y Leyendas

```python
"""
Se puede implementar estetica y legibilidad a los graficos añadiendo titulos,
etiquetas y leyendas.
"""

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.title("Raices cuadradras")
plt.plot(x_vals, y_vals, "r") # Podemos cambiar el color de la linea por ej: r = red
```

```python
"""
Para aniadir una leyenda hay que hacer dos cambios, pasarle un valor de string para el atributo label
y luego pasarle otro valor para el atributo loc del metodo de leyenda
"""

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.title("Raices cuadradras")
plt.plot(x_vals, y_vals, "r", label="Raiz cuadrada")
plt.legend(loc="upper center")
```

```python
"""
Tambien se le pueden pasar multiples lineas al diagrama dentro de un grafico
solo hay que llamar al metodo plot() varias veces con diferentes valores de x e y
"""

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
y2_vals = x_vals ** 3
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.title("Raices cuadradras")
plt.plot(x_vals, y_vals, "r", label="Raiz cuadrada")
plt.plot(x_vals, y2_vals, "b", label="Cubo")
plt.legend(loc="upper center")
```

## Graficos utilizando archivos CSV y TSV

```python
# leemos el dataset iris.csv, si no hay ningun error, es por que el archivo se cargo correctamente.

data = pd.read_csv("../datasets/iris_data.csv")
data.head()
```

```python
"""
Para trabajar con datasets, necesitamos pasarle los datos de la columna a la funcion plot() del modulo pyplot
para acceder al nombre de una columna desde pandas, necesitamos especificar el nombre del dataframe y en corchetes
la columna deseada.
"""

plt.xlabel("Largo del Sepalo")
plt.ylabel("Largo del Petalo")
plt.title("Largo del Sepalo vs Petalo")
plt.plot(data["sepal_length"], data["petal_length"], 'b')
```

```python
"""
Igual que a un CSV, se puede pasar un archivo TSV a traves del metodo read_csv(). solamente hay que pasarle '\t' en el valor
del parametro sep
"""

data = pd.read_csv("../datasets/iris_data.tsv", sep='\t')
data.head()
```

### Diagrama de dispersion

```python
"""
Los diagramas de dispersion son utilizados escencialmente para diagramar relaciones entre 2 columnas numericas, en forma
de puntos dispersos. normalmente un diagrama de dispersion es utilizado cuando para un mismo valor en el eje X existe muchos valores
en el eje Y. para este diagrama tenemos la funcion scatter()
"""

plt.xlabel("Largo del Sepalo")
plt.ylabel("Largo del Petalo")
plt.title("Largo del Sepalo vs Petalo")
plt.scatter(data['SepalLength'], data["PetalLength"], c="b")
```

### Grafico de barras

```python
"""
Los graficos de barras son utilizado para relaciones entre valores unicos en una columna categorica
agrupada por una funcion como suma, promedio, mediana, etc
"""

data = pd.read_csv("../datasets/titanic_data.csv")
data.head()
```

```python
"""
Para dibujar un grafico de barras, llamamos a la funcion bar() los valores categoricos se pasan en el eje x
y los valores numericos correspondientes en el eje Y
"""

plt.xlabel("Genero")
plt.ylabel("Edad")
plt.title("Genero VS Edad")
plt.bar(data["Sex"], data["Age"])
```

### Histogramas

```python
"""
Los histogramas son basicamente usados para mostrar la distribucion de datos para una lista numerica de items
"""

plt.title("Histograma de edad")
plt.hist(data["Age"])
```

### Graficos de Torta

```python
"""
Los graficos de torta como suguiere el nombre, muestra el porcentaje distribuido de valores en una columna categorica
para este tipo de graficos se utiliza la funcion pie()
"""

labels = "IT","Marketing","Data Science","Finance"
values = [500,156,300,510]
explode = (0.05,0.05,0.05,0.05) # Separacion entre partes

plt.pie(values, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show()
```

## Visualizacion utilizando Seaborn

```python
"""
Seaborn es una librearia construida arriba de matplotlib que nos facilita la visualizacion de datos como asi tambien
mejora el estilo de los mismos.
"""

plt.rcParams["figure.figsize"] = [10,8]
tips_data = sns.load_dataset("tips")
tips_data.head()

# El dataset de tips contiene registros de las cuentas pagadas por un cliente en un restaurante
```

### Diagrama distribucional

```python
"""
El diagrama distribucional es usado para mostrar un histograma de datos para una columna especifica en un dataset
para eso usamos la funcion displot()
"""

sns.displot(tips_data['total_bill'])
```

### Diagrama Joint Plot

```python
"""
Este diagrama se utiliza para datos bivariados, nos permite mostrar un histograma de distribucion de dos columnas
una en el eje X y otra en el eje Y para generar este grafico se utiliza la funcion jointplot()
"""

sns.jointplot(x='total_bill', y='tip', data=tips_data)
```

```python
# El scatter plot interno, puede ser remplazado por una linea de regresion, para hacerlo hay que pasarle el parametro reg

sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='reg')
```

### Grafico por pares

```python
"""
El grafico por pares es usado para crear un joint plot para todas las combinaciones de numeros y columas booleanas en un dataset
para esto se utiliza la funcion pairplot()
"""

sns.pairplot(data=tips_data)
```

### Grafico de barras SNS

```python
# Tambien podemos usar seaborn para mostrar graficos de barras

sns.set_style("darkgrid")
titanic_data = sns.load_dataset("titanic")
titanic_data.head()

sns.barplot(x="pclass", y="age", data=titanic_data) #La salida muestra el promedio de edad segun la clase.
```

### Diagrama de conteo

```python
"""
El diagrama de conteo es como el de barras, a diferencia del diagrama de barras que muestra los promedios, el count plot
simplemente muestra la cantidad de ocurrencias de cada registro por cada unico valor en la columna
"""

sns.countplot(x='pclass', data=titanic_data)
```

### Diagrama de caja

```python
"""
El diagrama de caja es usado para trazar informacion de un quartil para una columna de datos numericos, para utilizar este diagrama
necesitamos la funcion boxplot()
"""

sns.boxplot(x=titanic_data["fare"])
```

### Diagrama de Violin

```python
"""
Los diagramas de violin son similares a los de caja, pero a diferencia de los de caja que muestran los quartiles, el diagrama
de violin trazan la distribución general de valores en la columna numérica. Esto se aplica utilizando la funcion violinplot()
"""

sns.violinplot(x='alone', y='age', data=titanic_data)
```

## Visualizacion de datos con Pandas

```python
"""
Para comenzar con la visualizacion de datos con pandas, vamos a iniciar cargando un dataset
"""

titanic_data = pd.read_csv("../datasets/titanic_data.csv")
titanic_data.head()
```

```python
# Hay muchas formas de graficar con pandas, el primer acercamiento es seleccionando una columna y aplicar la funcion hist()


"""
Cuando utilizamos las funciones de los dataframes para graficar, tenemos que tener importado mathplotlib por que internamente
utiliza el modulo de pyplot
"""
titanic_data["Age"].hist()
```

```python
# Pandas Line Plots utlizando un dataset de vuelos

flights_data = sns.load_dataset('flights')
flights_data.head()
```

```python
flights_data.plot.line(y='passengers', figsize=(8,6))
```

```python
# El diagrama de dispersion con pandas tambien se utilizade una manera parecida.

flights_data.plot.scatter(x='year',y='passengers', figsize=(8,6))
```

```python
"""
Para crear un grafico de barras con pandas, es necesario una lista de categorias y una lista de valores
la lista de categorias y de valores deben tener el mismo largo
"""

# En este caso para mostrar el promedio de edad de los pasajeros femeninos y masculinos hay que calcularlo

titanic_data = pd.read_csv("../datasets/titanic_data.csv")
titanic_data.head()

sex_mean = titanic_data.groupby("Sex")["Age"].mean()

print(sex_mean)
print(type(sex_mean.tolist()))

df = pd.DataFrame({ "Gender": ["Female","Male"], "Age": sex_mean.tolist()})
ax = df.plot.bar(x="Gender", y="Age", figsize=(8,6))
```

```python
# Para crear box plot hay que utilizar la funcion box()

titanic_data.plot.box()
```
